```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import f_oneway
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_excel('DF_ayer23-2.xlsx')
df
```

```python
df.set_index("ID", inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.nunique()
```

```python
cols = ['Genotype', 'Developmental Stage']
df[cols] = df[cols].astype('category')

df.info()
```

```python
#Calculate correlation
#df = df.drop('ID', axis=1)
df.corr()
```

```python
sns.heatmap(df.corr(), cmap='coolwarm', annot=True)
plt.show()
```

```python
df.describe()
```

```python
# Calculate F-value and p-value using ANOVA
f_statistic, p_value = f_oneway(df['PND'], df['Filopodia'], df['Thin'], df['Longthin'], df['Mushroom'], df['Stubby'], df['Bifurcated'], df['Critical'], df['Non-critical']
                               ,df['Total'])
```

```python
print("F-value:", f_statistic)
print("p-value:", p_value)
```

```python
df.rename(columns={'Developmental Stage': 'Developmental_Stage'}, inplace=True)
```

```python
# Create an ANOVA model
model = ols("Filopodia ~ Genotype + PND + Developmental_Stage", data=df).fit()


# Perform ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
# Create a box plot to visualize Filopodia distribution by Genotype
plt.figure(figsize=(10, 6))
sns.boxplot(x="Genotype", y="Filopodia", data=df)
plt.title("Filopodia Distribution by Genotype")
plt.xlabel("Genotype")
plt.ylabel("Filopodia")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="PND", y="Filopodia", hue="Genotype", data=df)
plt.title("Scatter Plot of Filopodia vs. PND")
plt.xlabel("PND")
plt.ylabel("Filopodia")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Histogram
plt.subplot(2, 1, 1)
sns.histplot(data=df, x="Filopodia", kde=True)
plt.title("Histogram and Density Plot of Filopodia")
plt.xlabel("Filopodia")

# Density Plot
plt.subplot(2, 1, 2)
sns.kdeplot(data=df, x="Filopodia")
plt.xlabel("Filopodia")

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data=df, hue="Genotype")
plt.suptitle("Pairwise Relationships by Genotype")
plt.show()
```
